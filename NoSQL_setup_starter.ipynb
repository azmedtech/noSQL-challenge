4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
#Verify documents exist in the establishments collection for the Dover Locath Authority

# Find all unique LocalAuthorityName values in the establishments collection
unique_local_authorities = establishments.distinct("LocalAuthorityName")

# Print the unique LocalAuthorityName values
print("Unique Local Authority Names in the establishments collection:")
for authority in unique_local_authorities:
    print(authority)
```

```python
# Find how many documents have LocalAuthorityName as "Dover"
dover_query = {'LocalAuthorityName': 'Dover'}
dover_count = establishments.count_documents(dover_query)

print('Number of documents for establishments under the Dover local authority:', dover_count)
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
del_query = {'LocalAuthorityName': 'Dover'}
establishments.delete_many(del_query)
```

```python
# Check if any remaining documents include Dover
linger_query = {'LocalAuthorityName': 'Dover'}
linger_results = establishments.count_documents(linger_query)
linger_results
```

```python
# Check that other documents remain with 'find_one'
other_document = establishments.find_one()
pprint(other_document)
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
establishments.update_many({}, [{'$set':{"geocode.longitude": {'$toDouble': "$geocode.longitude"}, "geocode.latitude": {'$toDouble': "$geocode.latitude"}}}])
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many({}, [ {'$set':{ "RatingValue" : {'$toInt': "$RatingValue"}}} ])
```

```python
# Check that the coordinates and rating value are now numbers
check_fields = establishments.find_one()
pprint(check_fields)
```
